# Compute mean volume

```python
def get_mean_volume(symbol):
    df = pd.read_csv("data/{}.csv".format(symbol)) 
    return df['Volume'].mean()

def test_run():
    for symbol in ['AAPL', 'IBM']:
        print "Mean Volume"
        print symbol, get_mean_volume(symbol)

test_run()
```
